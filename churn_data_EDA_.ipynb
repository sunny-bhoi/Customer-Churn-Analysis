```python
#from chat gpt


# List of columns you want to plot
cols = [
    'Phone_Service', 'Multiple_Lines', 'Internet_Service', 'Internet_Type',
    'Online_Security', 'Online_Backup', 'Device_Protection_Plan',
    'Premium_Support', 'Streaming_Movies', 'Streaming_Music',
    'Unlimited_Data', 'Paperless_Billing', 'Payment_Method', 'Streaming_TV '
]

# Set up the grid layout: 5 rows x 3 columns (adjust if needed)
rows = 5
cols_per_row = 3
fig, axes = plt.subplots(nrows=rows, ncols=cols_per_row, figsize=(18, 20))
axes = axes.flatten()  # Flatten to make it easy to loop

# Column list (you already defined above)
columns_to_plot = [
    'Phone_Service', 'Multiple_Lines', 'Internet_Service', 'Internet_Type',
    'Online_Security', 'Online_Backup', 'Device_Protection_Plan',
    'Premium_Support', 'Streaming_Movies', 'Streaming_Music',
    'Unlimited_Data', 'Paperless_Billing', 'Payment_Method', 'Streaming_TV '
]

# Loop through each column and plot
for i, col in enumerate(columns_to_plot):
    sns.countplot(data=df, x=col, hue="Customer_Status", ax=axes[i], palette="pastel")
    axes[i].set_title(f'{col} vs Customer_Status', fontsize=11)
    axes[i].tick_params(axis='x', rotation=30)

# Turn off any unused axes (if less than 15 plots)
for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Based on the countplots:

# High Retention Services:
# Customers with services like Phone Service, Internet Service, Online Backup, and Unlimited Data show significantly higher retention.
# These offerings may be essential for customer satisfaction and loyalty.

# Security & Support Matter:
# Users with Online Security, Premium Support, and Device Protection Plans are more likely to stay. 
# Their absence correlates with higher churn, highlighting the importance of perceived service protection and value.

# Streaming Services Show Mixed Signals:
# Streaming TV and Streaming Music users are relatively balanced across all statuses. 
# While they’re widely used, they don’t appear to be strong retention drivers alone.

# Billing Preferences Influence Retention:
# Customers opting for Paperless Billing and Credit Card or Bank Withdrawal payments show better retention than those using Mailed Check,
# which sees higher churn. Convenience-oriented billing may support retention.

# Churn Signals:
# Churn is higher among users who lack core digital services and customer support options. In contrast, those with bundled features 
# (e.g., multiple lines, device protection, unlimited data) tend to stay longer.
```
